```python
df_subject = pd.read_csv(f'/home/weiyanpeh/Git/SFM_Related/CADENCE/SHAPE/subject_summary_video_join.csv')

columns = [
    'Subject ID', 'Visit ID', 
    'Videos', 'video_date_LR',
    'Weight', 'BMI', 'HR', 'SPO2',
    'Left foot length (cm)', 'Left foot width (cm)',
    'Left foot circumference at center (cm)',
    'Left foot ankle circumference (cm)', 'Right foot length (cm)',
    'Right foot width (cm)', 'Right foot circumference at center (cm)',
    'Right foot ankle circumference (cm)',
    ]

df_subject = df_subject[columns]
df_subject
```

```python
def convert_and_replace_nan(x):
    if isinstance(x, str):
        try:
            val = ast.literal_eval(x)
            if isinstance(val, list):
                return [0 if isinstance(i, str) and i.strip().lower() == 'nan' else i for i in val]
        except (ValueError, SyntaxError):
            return x
    return 0 if pd.isna(x) else x




for col in columns:
    # print(col)
    if col != 'Subject ID':
        # Convert string to list only if it contains one item
        df_subject[col] = df_subject[col].apply(lambda x: x.replace('nan', '0'))
        df_subject[col] = df_subject[col].apply(lambda x: convert_and_replace_nan(x))

def clean_mp(x):
    for i in range(len(x)):
        x[i] = x[i].replace('.mp4', '')
    return x

df_subject['Videos'] = df_subject['Videos'].apply(lambda x: clean_mp(x))

df_subject
```

```python
# df = pd.read_csv(f'/home/weiyanpeh/Git/SFM_Related/CADENCE/results_3_final_volume_perfect.csv')
df = pd.read_csv(f'/home/weiyanpeh/Git/SFM_Related/CADENCE/results_3_final_volume.csv')
df = df.drop(['ply_path', 'file_name'], axis=1)

# df = df[['subject', 'data_image', 'feet_length', 'volume', 'vol_est']]
df = df[['subject', 'data_image', 'feet_length', 'volume']]
df = df[df['volume'] != 0].reset_index(drop=True)
df = df[df['subject'] != 74].reset_index(drop=True)
df
```

```python
# list_patient = [1, 4, 6, 7, 12, 15, 17, 19, 29, 35, 46, 74]
list_patient = list(set(df['subject'].tolist()))
for patient in list_patient:
    print('#'*60)
    print(f'Subject {patient}')
    df_temp = df[df['subject'] == patient]

    list_data_image = list(df_temp['data_image'])

    list_position = []

    list_l = []
    list_r = []


    try:
    # if True:
        for data_image in list_data_image:
            index = int(data_image.split('D')[1].replace('R', '').replace('L', ''))
            # print(f'-> {data_image} - {index}')
            
            df_data_temp = df_subject[df_subject['Subject ID'] == patient].reset_index(drop=True)
            list_index = list(df_data_temp['Visit ID'][0])
            position = list_index.index(index)
            
            volume = df_temp[df_temp['data_image'] == data_image]['volume'].reset_index(drop=True)[0]
            # volume = volume/100
            

            if 'L' in data_image:
                column = 'Left'
            else:
                column = 'Right'
                
            columns = [
                f'{column} foot ankle circumference (cm)',
                f'{column} foot circumference at center (cm)',
                # f'{column} foot length (cm)',
                # f'{column} foot width (cm)'
            ]
            
            list_temp = []
            for col in columns:
                value = list(df_data_temp[col][0])[position]
                # print(f'{col} - {value}')
                
                list_temp.append(value)
                
            if column == 'Left':
                list_l.append(list_temp + [volume])

            else:
                list_r.append(list_temp + [volume])
                
            
        def process_df(list_feet):
            columns_save = ['ankle_c', 'feet_c', 'volume']
            df_feet = pd.DataFrame(list_feet, columns=columns_save)
            df_feet['average'] = df_feet[['ankle_c', 'feet_c']].mean(axis=1)
            
            df_feet['pct_change'] = df_feet['volume'].pct_change()
            # df_feet['pct_change'].iloc[0] = 0
            df_feet.loc[0, "pct_change"]
            df_feet['pct_change'] = df_feet['pct_change'] * 100
            df_feet['direction'] = df_feet['volume'].diff().apply(lambda x: 1 if x > 0 else (-1 if x < 0 else 0))
            
            # Create a time index
            df_feet.index = pd.Index(range(1, len(df_feet)+1), name='Time')
        
            return df_feet
        
        df_l = process_df(list_l)
        df_r = process_df(list_r)
        
        print('Left:', df_l['volume'])
        print('Right:', df_r['volume'])

        try:
            print('Left:', df_l['vol_est'])
            print('Right:', df_r['vol_est'])
        except:
            pass

        # Plotting side-by-side
        fig, axs = plt.subplots(2, 2, figsize=(8, 5), sharex=True)

        # Right foot subplot
        for column in ['ankle_c', 'feet_c', 'average']:
            axs[0, 0].plot(df_l.index, df_l[column], label=column, marker='o')
        axs[0, 0].set_title('Left Foot Measurements')
        axs[0, 0].set_ylabel('Value')
        axs[0, 0].legend()

        for column in ['direction']:
            axs[1, 0].plot(df_l.index, df_l[column], label=column, marker='o')
        axs[1, 0].set_title('Left Foot Volume')
        axs[1, 0].set_xlabel('Day')
        axs[1, 0].set_ylabel('Direction (+ / -)')
        axs[1, 0].set_ylim(-1.2, 1.2)

        # Left foot subplot
        for column in ['ankle_c', 'feet_c', 'average']:
            axs[0, 1].plot(df_r.index, df_r[column], label=column, marker='o')
        axs[0, 1].set_title('Right Foot Measurements')
        axs[0, 1].legend()
        
        for column in ['direction']:
            axs[1, 1].plot(df_r.index, df_r[column], label=column, marker='o')
        axs[1, 1].set_title('Right Foot Volume')
        axs[1, 1].set_xlabel('Days')
        axs[1, 1].set_ylabel('Direction (+ / -)')
        axs[1, 1].set_ylim(-1.2, 1.2)
        
        plt.suptitle(f'Subject:{patient}')

        plt.tight_layout()
        plt.show()
        
    except Exception as error:
        print(error)

            
```
